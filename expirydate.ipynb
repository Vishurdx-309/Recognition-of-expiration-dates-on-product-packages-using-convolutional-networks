```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
!pip install matplotlib
import matplotlib.pyplot as plt
```

```python
x_train = tf.keras.utils.normalize(x_train,axis = 1)
x_test = tf.keras.utils.normalize(x_test,axis = 1)
plt.imshow(x_train[0],cmap = plt.cm.binary)
```

```python
!pip install numpy
import numpy as np
```

```python
IMG_SIZE = 28
x_trainr = np.array(x_train).reshape(-1,IMG_SIZE,IMG_SIZE,1)
x_testr = np.array(x_test).reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D
```

```python
model = Sequential()

model.add(Conv2D(64, (3,3), input_shape = x_trainr.shape[1:] ))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(64, (3,3), input_shape = x_trainr.shape[1:] ))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(64, (3,3), input_shape = x_trainr.shape[1:] ))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))

model.add(Dense(32))
model.add(Activation("relu"))

model.add(Dense(10))
model.add(Activation("softmax"))
```

```python
model.summary()
```

```python
model.compile(loss ="sparse_categorical_crossentropy" , optimizer  = "adam" ,metrics = ['accuracy'])
```

```python
model.fit(x_trainr,y_train,epochs = 10, validation_split= 0.3)
```

```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
from PIL import Image
import pytesseract
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

model_date = YOLO('/content/drive/MyDrive/colabweights/date.pt')
model_numbers = YOLO('/content/drive/MyDrive/colabweights/numbers.pt')
def extract_date_components(detected_numbers_and_slash):
    components = detected_numbers_and_slash.split('/')
    year = int(components[2]) if len(components) > 2 else None
    month = int(components[0]) if len(components) > 0 else None
    day = int(components[1]) if len(components) > 1 else None

    return  year, month, day
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    _, thresholded = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return thresholded
def detect_numbers_in_image(image_path):

    preprocessed_image = preprocess_image(image_path)

    pil_image = Image.fromarray(preprocessed_image)

    text = pytesseract.image_to_string(pil_image)

    numbers_and_slash = ''.join(c if c.isdigit() or c == '/' else ' ' for c in text)

    return numbers_and_slash.strip()

frame = cv2.imread('/content/drive/MyDrive/colabweights/eximage.webp')

results_date = model_date(frame, save=False, show=False)

for r_date in results_date:
    boxes_date = r_date.boxes.xyxy.cpu().numpy()

    frame_with_boxes_date = frame.copy()
    for box_date in boxes_date:
        print("DATE FOUND")
        cv2.rectangle(frame_with_boxes_date, (int(box_date[0]), int(box_date[1])), (int(box_date[2]), int(box_date[3])), (0, 0, 255), 2)


        cropped_image = frame[int(box_date[1]):int(box_date[3]), int(box_date[0]):int(box_date[2])]
        cv2.imwrite('/content/drive/MyDrive/colabweights/croppedimage.jpg', cropped_image)

        croppedimage ='/content/drive/MyDrive/colabweights/croppedimage.jpg'

        detected_numbers = detect_numbers_in_image('/content/drive/MyDrive/colabweights/croppedimage.jpg')

        if detected_numbers:
          print(f"Detected Date: {detected_numbers}")
          date = detected_numbers
          print(extract_date_components(date))

        else:
          print("No numbers detected.")

        cv2_imshow(cropped_image)

        results = model_numbers(croppedimage, save=False, show=False)

        for result in results:
            boxes_numbers = result.boxes.xyxy.cpu().numpy()
            probs = result.probs
            frame_with_numbers = cropped_image.copy()
            for box in boxes_numbers:

                print("Number FOUND")

            cv2_imshow(frame_with_numbers)

cv2_imshow(frame_with_boxes_date)
cv2.waitKey(0)
```
